```python
price_list_by_n=[price_sub1,price_sub2,price_sub3,price_sub4,price_sub5]
```

```python
p_l_b_n_2=[]
neigh_list=['Brooklyn','Manhattan','Queens','State n Island','Bronx']
for x in price_list_by_n:
  i=x.describe(percentiles=[.25,.50,.75])
  i=i.iloc[3:]
  i.reset_index(inplace=True)
  i.rename(columns={'index':'Stats'},inplace=True)
  p_l_b_n_2.append(i)
```

```python
p_l_b_n_2[0].rename(columns={'price':neigh_list[0]}, inplace=True)
p_l_b_n_2[1].rename(columns={'price':neigh_list[1]}, inplace=True)
p_l_b_n_2[2].rename(columns={'price':neigh_list[2]}, inplace=True)
p_l_b_n_2[3].rename(columns={'price':neigh_list[3]}, inplace=True)
p_l_b_n_2[4].rename(columns={'price':neigh_list[4]}, inplace=True)
```

```python
stat_df=p_l_b_n_2
stat_df=[df.set_index('Stats') for df in stat_df]
stat_df=stat_df[0].join(stat_df[1:])
stat_df
```

```python
_names_=[]
for name in airbnb.name:
    _names_.append(name)  
def split_name(name):
    spl=str(name).split()
    return spl
_names_for_count_=[]
for x in _names_:
    for word in split_name(x):
        word=word.lower()
        _names_for_count_.append(word)
```

```python
from collections import Counter
_top_25_w=Counter(_names_for_count_).most_common()
_top_25_w=_top_25_w[0:25]
```

```python
sub_w=pd.DataFrame(_top_25_w)
sub_w.rename(columns={0:'Words', 1:'Count'}, inplace=True)
```

```python
bnb_5=sns.barplot(x='Words', y='Count', data=sub_w)
bnb_5.set_title('Counts of the top 25 used words for listing names')
bnb_5.set_ylabel('Count of words')
bnb_5.set_xlabel('Words')
bnb_5.set_xticklabels(bnb_5.get_xticklabels(), rotation=80)
```
